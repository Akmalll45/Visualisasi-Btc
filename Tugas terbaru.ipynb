```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
# Generate 'random' data
np.random.seed(0)
X = 2.5 * np.random.randn(100) + 1.5 # Array of 100 values with mean = 1.5, stddev = 2.5
res = 0.5 * np.random.randn(100) # Generate 100 residual terms
y = 2 + 0.3 * X + res # Actual values of Y
# Create pandas dataframe to store our X and y values
df = pd.DataFrame(
{'X': X,
'y': y}
)
# Show the first five rows of our dataframe
df.head()
```

```python
# Calculate the mean of X and y
xmean = np.mean(X)
ymean = np.mean(y)
# Calculate the terms needed for the numator and denominator of beta
df['xycov'] = (df['X'] - xmean) * (df['y'] - ymean)
df['xvar'] = (df['X'] - xmean)**2
# Calculate beta and alpha
beta = df['xycov'].sum() / df['xvar'].sum()
alpha = ymean - (beta * xmean)
print(f'alpha = {alpha}')
print(f'beta = {beta}')
ypred = alpha + beta * X # fungsi prediksi
# Plot regression against actual data
plt.figure(figsize=(12, 6))
plt.plot(X, ypred) # regression line
plt.plot(X, y, 'ro') # scatter plot showing actual data
plt.title('Actual vs Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import statsmodels.api as sm
# Menghasilkan data acak
np.random.seed(42)
x1 = np.random.randint(1, 51, 50) # Angka acak untuk x1
x2 = np.random.randint(3, 61, 50) # Angka acak untuk x2
y = 10 + 2 * x1 + 3 * x2 + np.random.normal(0, 10, 50) # Hubungan linear dengan noise
# Membuat DataFrame
df = pd.DataFrame({
"x1": x1,
"x2": x2,
"y": y
})
# Analisis regresi berganda
X = sm.add_constant(df[["x1", "x2"]]) # Menambahkan konstanta untuk intersep
model = sm.OLS(df["y"], X).fit() # Fit model regresi
# Mendapatkan koefisien regresi
koefisien = model.params
# Scatter plot 3D
scatter_plot = go.Scatter3d(
x=df["x1"],
y=df["x2"],
z=df["y"],
mode="markers",
marker=dict(
size=5,
color="blue",
opacity=0.8
),
name="Data"
)
# Membuat grid untuk permukaan regresi
x1_grid, x2_grid = np.meshgrid(np.linspace(1, 50, 10), np.linspace(3, 60, 10))
y_pred = koefisien[0] + koefisien[1] * x1_grid + koefisien[2] * x2_grid
# Permukaan regresi 3D
regression_surface = go.Surface(
x=x1_grid,
y=x2_grid,
z=y_pred,
colorscale="Reds",
opacity=0.5,
name="Garis Regresi"
)
# Membuat layout plot
layout = go.Layout(
title="Scatter Plot 3D dengan Garis Regresi Berganda",
scene=dict(
xaxis_title="x1",
yaxis_title="x2",
zaxis_title="y"
),
showlegend=True
)
# Membuat figure dengan scatter plot dan permukaan regresi
fig = go.Figure(data=[scatter_plot, regression_surface], layout=layout)
# Menampilkan plot
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')

#  Ambil data harga Bitcoin dari GitHub
print("ðŸ“Š Loading Bitcoin data...")
url = "https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/BTC-USD.csv"
df = pd.read_csv(url)

# ðŸ—“ï¸ Ubah kolom Date jadi format tanggal, terus dipakai jadi index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

#  Bikin indikator tambahan biar datanya lebih kaya
df['High_Low_Ratio'] = df['High'] / df['Low']     # bandingin harga tertinggi sama terendah
df['Price_Range'] = df['High'] - df['Low']       # selisih harga tertinggi dan terendah
df = df.dropna()

# ðŸŽ¯ Tentuin target (harga penutupan) dan fitur yang dipakai
target = 'Close'
selected_features = ['Volume', 'High_Low_Ratio', 'Price_Range']

#  Pisahin data fitur (X) sama target (y)
X_data = df[selected_features]
y_data = df[target]

#  Normalisasi fitur biar skalanya sama rata
scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X_data), 
    columns=X_data.columns, 
    index=X_data.index
)

#  Bikin model regresi OLS (kayak regresi linear tapi versi statistik)
X_with_const = sm.add_constant(X_scaled)
model = sm.OLS(y_data, X_with_const).fit()
y_pred = model.predict(X_with_const)
coefficients = model.params

#  Siapin data buat plot 3D
feature_x = selected_features[0]  # Volume
feature_y = selected_features[1]  # High_Low_Ratio

x_vals = X_scaled[feature_x]
y_vals = X_scaled[feature_y]
z_vals = y_data

#  Scatter plot 3D buat nampilin data asli Bitcoin
scatter_3d = go.Scatter3d(
    x=x_vals, 
    y=y_vals, 
    z=z_vals,
    mode='markers',
    text=df.index.strftime('%Y-%m-%d'),  # kalau di-hover keluar tanggal
    hovertemplate=(
        " Tanggal: %{text}<br>" +
        " Volume (norm): %{x:.2f}<br>" +
        " HL Ratio (norm): %{y:.2f}<br>" +
        " Close: $%{z:,.0f}<extra></extra>"
    ),
    marker=dict(
        size=4,
        color=z_vals,
        colorscale='Hot',
        opacity=0.8,
        line=dict(width=0.5, color='DarkSlateGrey'),
        colorbar=dict(
            title=" Harga ($)",
            tickformat="$,.0f"
        )
    ),
    name='Data Bitcoin'
)

#  Bikin permukaan (surface) dari hasil regresi biar kelihatan trend garisnya
x_range = np.linspace(x_vals.min(), x_vals.max(), 25)
y_range = np.linspace(y_vals.min(), y_vals.max(), 25)
x_grid, y_grid = np.meshgrid(x_range, y_range)

other_features_mean = X_scaled[selected_features[2:]].mean().values if len(selected_features) > 2 else []
z_surface = coefficients[0] + coefficients[1]*x_grid + coefficients[2]*y_grid
if len(selected_features) > 2:
    for i, mean_val in enumerate(other_features_mean):
        z_surface += coefficients[i+3]*mean_val

surface_3d = go.Surface(
    x=x_grid, 
    y=y_grid, 
    z=z_surface,
    colorscale='Reds',
    opacity=0.4,
    showscale=False,
    name='Regression Surface'
)

#  Atur layout plot
fig = go.Figure(data=[scatter_3d, surface_3d])
fig.update_layout(
    title=f"<b>Regresi Harga Bitcoin (3D)</b>",
    scene=dict(
        xaxis_title=f" {feature_x} (Normalized)",
        yaxis_title=f" {feature_y} (Normalized)",
        zaxis_title=" Harga ($)"
    ),
    height=700,
    margin=dict(l=50, r=50, t=80, b=50)
)

# ðŸš€ Tampilkan hasil plot interaktif
fig.show()
```
